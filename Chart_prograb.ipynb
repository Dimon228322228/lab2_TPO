```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Считываем все файлы
complex_df = pd.read_csv('csv_complex.csv')
cos_df = pd.read_csv('csv_cos.csv')
cot_df = pd.read_csv('csv_cot.csv')
csc_df = pd.read_csv('csv_csc.csv')
ln_df = pd.read_csv('csv_ln.csv')
log2_df = pd.read_csv('csv_log_2.0.csv')
log10_df = pd.read_csv('csv_log_10.0.csv')
sec_df = pd.read_csv('csv_sec.csv')
sin_df = pd.read_csv('csv_sin.csv')
tg_df = pd.read_csv('csv_tg.csv')
```

```python
# функция для построения графика
def plot_function(df, title):
    x = df['X']
    f_x = df['f(X)']
    plt.figure(figsize=(10, 6))
    plt.plot(x, f_x, marker='o', linestyle='-', color='blue', label='f(x)')

    # Настраиваем график
    plt.title(title, fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.savefig(f'charts/{title}.png', dpi=300, bbox_inches='tight')

    # Отображаем график
    plt.tight_layout()
    plt.show()
```

```python
# Построение графиков

plot_function(complex_df, 'График функции complex')
plot_function(cos_df, 'График функции cos')
plot_function(cot_df, 'График функции cot')
plot_function(csc_df, 'График функции csc')
plot_function(ln_df, 'График функции ln')
plot_function(log2_df, 'График функции log_2.0')
plot_function(log10_df, 'График функции log_10.0')
plot_function(sec_df, 'График функции sec')
plot_function(sin_df, 'График функции sin')
plot_function(tg_df, 'График функции tg')
```
